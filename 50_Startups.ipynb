```python
import pandas as pd
import numpy as np
```

Reading a CSV File

```python
startups = pd.read_csv("50_Startups.csv")
```

Handling Missing Values

```python
startups.head()
```

```python
startups.isnull()
```

```python
startups.isnull().sum()
```

But there are 0 values present in the database which represents the NUll value

```python
startups.replace(0,np.nan,inplace = True)
```

```python
startups.tail()
```

```python
startups.isnull().sum()
```

Replacing the Missing values with their mean values

```python
avg_rd_spend = startups['R&D Spend'].mean(axis=0)
```

```python
avg_marketing_spend = startups['Marketing Spend'].mean(axis = 0)
```

```python
startups['R&D Spend'].replace(np.nan,avg_rd_spend,inplace = True)
```

```python
startups['Marketing Spend'].replace(np.nan,avg_marketing_spend,inplace = True)
```

```python
startups.tail()
```

Now, We do OneHotEncoder for the State column in the database using Pandas

```python
database_encoder = pd.get_dummies(startups['State'])
```

```python
database_encoder.head()
```

Now, we concat this database in the original database

```python
startups = pd.concat([startups,database_encoder],axis=1)
```

```python
startups.head(10)
```

We drop the State column from the database

```python
startups.drop('State',axis = 1, inplace = True)
```

```python
startups.head(10)
```

Now, our database is preprocessed.
