```python
import pandas as pd
import numpy as np
```

Reading a CSV File

```python
startups = pd.read_csv("50_Startups.csv")
```

```python
startups.head()
```

Now, We do OneHotEncoder for the State column in the database using Pandas

```python
database_encoder = pd.get_dummies(startups['State'])
```

```python
database_encoder.head()
```

Now, we concat this database in the original database

```python
startups = pd.concat([startups,database_encoder],axis=1)
```

```python
startups.head(10)
```

We drop the State column from the database

```python
startups.drop('State',axis = 1, inplace = True)
```

```python
startups.head(10)
```

Now, our database is preprocessed.
